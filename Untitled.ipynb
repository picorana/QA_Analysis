```python
import json
import traceback

data = []

no_asin = set()
yes_asin = set()

data = {}

output = open("output.json", 'w')

with open('Cell_Phones_and_Accessories_5.json', 'r') as f:
    
    for i, line in enumerate(f):
        
        if i%100==0: print i, len(data)
        
        d = json.loads(line)
        if d['asin'] in no_asin: continue
            
        if d['asin'] in yes_asin: 
            data[d['asin']]['reviewTexts'].append(d['reviewText'])
            
        try:
            product = amazon.lookup(ItemId=d['asin'])
            
            yes_asin.add(d['asin'])
                           
            this_dict = {}
     
            this_dict['features'] = product.features
            this_dict['brand'] = product.brand
            this_dict['label'] = product.label

            this_dict['reviewTexts'] = []
            this_dict['reviewTexts'].append(d['reviewText'])
                
            data[d['asin']] = this_dict

        except Exception as e:
            print traceback.format_exc()
            no_asin.add(d['asin'])
```
