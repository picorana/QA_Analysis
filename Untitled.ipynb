```python
def add_asins_to_examine(search_term):
    
    check = False
    
    while check == False:
        
        time.sleep(3)
        
        try:
            products = amazon.search(Keywords = search_term, SearchIndex='All')

            asins_to_examine = open("asins_to_examine.txt", 'r')
            examined_asins = open("examined_asins.txt", 'r')
            asins_with_problems = open("asins_with_problems.txt", 'r')

            already_added = set()

            for line in asins_to_examine:
                already_added.add(line)
            for line in examined_asins:
                already_added.add(line)
            for line in asins_with_problems:
                already_added.add(line)

            asins_to_examine = open("asins_to_examine.txt", 'a')

            count = 0
            for product in products:
                print product.asin
                count += 1
                if product not in already_added:
                    asins_to_examine.write(product.asin + '\n')
            check = True
            
            print "found " + str(count) + " products for search term " + search_term
            
        except Exception as e:
            print e
            print 'trying again term ' + search_term
            
for word in search_terms:
    add_asins_to_examine(word)
```

```python
import json
import time
import os
import traceback
import logging

f = open("asins_to_examine.txt", 'r')

done_list = set()
for x in os.listdir("product_data"):
    done_list.add(x[:-5])
    
    
for i, line in enumerate(f):

    if line.strip() in done_list: continue
        
    check = False

    while check == False:
        
        time.sleep(3)
        
        try:
            print "trying " + line.strip()
            product = amazon.lookup(ItemId=line.strip())

            this_dict = get_product_dict(product)
            
            if "price_and_currency" in this_dict:
                #print this_dict["price_and_currency"]
                if this_dict["price_and_currency"] != (None, None):
                    this_dict["price_and_currency"] = float(product.price_and_currency[0])
                else: this_dict["price_and_currency"] = ""
                    
            if "release_date" in this_dict and this_dict["release_date"] != None:
                this_dict["release_date"] = this_dict["release_date"].strftime("%Y-%m-%d %H:%M:%S")
                
            if "publication_date" in this_dict and this_dict["publication_date"] != None:
                this_dict["publication_date"] = this_dict["publication_date"].strftime("%Y-%m-%d %H:%M:%S")
            
            if "formatted_price" in this_dict:
                #print this_dict["formatted_price"]
                if this_dict["formatted_price"] != None:
                    this_dict["formatted_price"] = float(product.formatted_price[1:])
                else:
                    this_dict["formatted_price"] = ""
            
            #pprint(this_dict)
            #if "detail_page_url" in this_dict:
            #    this_dict["detail_page_url"] = this_dict["detail_page_url"].split('?')[0]
                
            f_res = open("product_data/"+this_dict['asin']+".json", 'w+')

            json.dump(this_dict, f_res, indent=4)

            check = True
        
        except Exception as e:
            logging.error(traceback.format_exc())
            if "AsinNotFound" in traceback.format_exc(): check = True
            print 'trying again ' + line.strip()
            
         
```
