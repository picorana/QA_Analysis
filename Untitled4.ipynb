```python
def retrieve_info_from_page(asin):

    item_characteristics_dict = {}

    page = open('./pages/'+asin, 'r')
    soup = BeautifulSoup(page.read(), 'html.parser')
    tab = soup.find('div', {'id': 'HLCXComparisonWidget_feature_div'})

    item_characteristics_dict['name'] = soup.find('span', {'id':'productTitle'}).text.strip()
    item_characteristics_dict['asin'] = asin
    
    bullets = soup.find('div', {'id':'feature-bullets'})
    #for span in bullets.findAll('span'):
        #print span.text.strip()

    try:
        for i, tr in enumerate(tab.findAll('tr')):
            # tr 0:
            # tr 1:
            # tr 2:
            # tr 3: customer rating
            if i<4: continue

            key = tr.findAll('td')[0].span.text
            if tr.findAll('td')[1].span != None: value = tr.findAll('td')[1].span.text
            else: value = tr.findAll('td')[1].a.text
            item_characteristics_dict[key.strip().lower()] = value.strip().lower()
    except: 
        print 'no tr tab'
        
    # PROD DETAILS
    details_tab = soup.find('div', {'id':'prodDetails'})
    #print details_tab

    if soup.find('div', {'id':'productDescription'})!=None: 
        description1 = soup.find('div', {'id':'productDescription'}).p.text
        print description1

    pprint(item_characteristics_dict, indent=4)
    
    review_list = []
    f = open('./results/' + asin[:-5] + '.json', 'r')
    for line in f:
        review_list.append(json.loads(line)['text'])
        
    rel_list = []
    for r in review_list:
        s = parsetree(r.encode('ascii', 'ignore'), relations=True, lemmata=True)
        for sentence in s:
            rel_dict = {}
        
            for rel in sentence.relations:
                for rnum in sentence.relations[rel]:

                    if rnum not in rel_dict: 
                        rel_dict[rnum] = {}
                        rel_dict[rnum][rel] = sentence.relations[rel][rnum]
                    else:
                        rel_dict[rnum][rel] = sentence.relations[rel][rnum]

            rel_list.append(rel_dict)
    
    return item_characteristics_dict, rel_list

```

```python
train_file = open('qa_Cell_Phones_and_Accessories.json', 'r')

list_of_questions_dicts = []

for i, line in enumerate(train_file):
    line = line.replace('"', "'")
    line = line.replace("{'", '{"')
    line = line.replace("':", '":')
    line = line.replace(": '", ': "')
    line = line.replace("',", '",')
    line = line.replace(", '", ', "')
    line = line.replace("'}", '"}')
    line = line.replace("\\", "")
    
    try:
        list_of_questions_dicts.append(json.loads(line))
    except Exception as e:
        continue
```

```python
list_of_questions = []
list_of_answers = []
list_of_labels = []
list_of_labels_answers = []

wordnet_lemmatizer = WordNetLemmatizer()

for i, quest in enumerate(list_of_questions_dicts):
    
    # QUESTIONS
    qstring = ''.join([i for i in quest['question'] if not i.isdigit()])
    qstring = [x.strip(string.punctuation) for x in qstring.split()]
    
    new_list = []
    for j, word in enumerate(qstring):
        new_list.append(wordnet_lemmatizer.lemmatize(word.lower()))
        
    list_of_questions.append(new_list)
    
    if quest['questionType'] == "yes/no": list_of_labels.append(1)
    else: list_of_labels.append(0)

    # ANSWERS
    if 'answerType' not in quest:
        continue
    
    if quest['answerType'] == '?':
        continue
    
    for i in quest:
        if i=='answerType': 
            list_of_labels_answers.append(quest[i])
    
    astring = ''.join([i for i in quest['answer'] if not i.isdigit()])
    astring = [x.strip(string.punctuation) for x in astring.split()]
    
    new_list = []
    for j, word in enumerate(astring):
        new_list.append(wordnet_lemmatizer.lemmatize(word.lower()))
    list_of_answers.append(new_list)

    
```

```python
def retrieve_word_features_from_lists_of_tokens(working_list, is_answers=False):
    
    all_words_tmp = []
    
    if is_answers:
        for l in working_list:
            for word in l: all_words_tmp.append(word)
    else:    
        for l in working_list:
            for g in nltk.ngrams(l, 2):
                all_words_tmp.append(g)
            for g in nltk.ngrams(l, 3):
                all_words_tmp.append(g)

    all_words = nltk.FreqDist(all_words_tmp)

    print len(all_words)

    word_features = []
    for s in all_words:
        if all_words[s] > 10: word_features.append(s)
        
    print len(word_features)
    return word_features
    


def document_features(document, word_features, is_answers=False):
    
    qstring = ' '.join([i for i in document if not i.isdigit()])
    document = [x.strip(string.punctuation) for x in qstring.split()]
    
    if is_answers:
        document_words = set(document)
    else:
        document_words = set(nltk.ngrams(document, 2))
        document_words = document_words.union(set(nltk.ngrams(document, 3)))
    
    features = {}
    check = 0
    for word in word_features:
        if word in document_words:
            features['contains({})'.format(word)] = True
            check = 1
        else:
            features['contains({})'.format(word)] = False
    return features, check

word_features_questions = retrieve_word_features_from_lists_of_tokens(list_of_answers)
word_features_answers = retrieve_word_features_from_lists_of_tokens(list_of_answers, is_answers=True)
```

```python
featuresets = []
for i in range(len(list_of_questions)/50):
    res, check = document_features(list_of_questions[i], word_features_questions)
    if check != 0: featuresets.append((res, list_of_labels[i]))
```

```python
train_set, test_set = featuresets[len(featuresets)*7/10:], featuresets[:-len(featuresets)*3/10]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```
