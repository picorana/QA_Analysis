```python
def retrieve_info_from_page(asin):

    item_characteristics_dict = {}

    page = open('./pages/'+asin, 'r')
    soup = BeautifulSoup(page.read(), 'html.parser')
    tab = soup.find('div', {'id': 'HLCXComparisonWidget_feature_div'})

    item_characteristics_dict['name'] = soup.find('span', {'id':'productTitle'}).text.strip()
    item_characteristics_dict['asin'] = asin
    
    bullets = soup.find('div', {'id':'feature-bullets'})
    #for span in bullets.findAll('span'):
        #print span.text.strip()

    try:
        for i, tr in enumerate(tab.findAll('tr')):
            # tr 0:
            # tr 1:
            # tr 2:
            # tr 3: customer rating
            if i<4: continue

            key = tr.findAll('td')[0].span.text
            if tr.findAll('td')[1].span != None: value = tr.findAll('td')[1].span.text
            else: value = tr.findAll('td')[1].a.text
            item_characteristics_dict[key.strip().lower()] = value.strip().lower()
    except: 
        print 'no tr tab'
        
    # PROD DETAILS
    details_tab = soup.find('div', {'id':'prodDetails'})
    #print details_tab

    if soup.find('div', {'id':'productDescription'})!=None: 
        description1 = soup.find('div', {'id':'productDescription'}).p.text
        print description1

    pprint(item_characteristics_dict, indent=4)
    
    review_list = []
    f = open('./results/' + asin[:-5] + '.json', 'r')
    for line in f:
        review_list.append(json.loads(line)['text'])
        
    rel_list = []
    for r in review_list:
        s = parsetree(r.encode('ascii', 'ignore'), relations=True, lemmata=True)
        for sentence in s:
            rel_dict = {}
        
            for rel in sentence.relations:
                for rnum in sentence.relations[rel]:

                    if rnum not in rel_dict: 
                        rel_dict[rnum] = {}
                        rel_dict[rnum][rel] = sentence.relations[rel][rnum]
                    else:
                        rel_dict[rnum][rel] = sentence.relations[rel][rnum]

            rel_list.append(rel_dict)
    
    return item_characteristics_dict, rel_list

```

```python
train_file = open('qa_Pet_Supplies.json', 'r')

list_of_questions_dicts = []

for i, line in enumerate(train_file):
    line = line.replace('"', "'")
    line = line.replace("{'", '{"')
    line = line.replace("':", '":')
    line = line.replace(": '", ': "')
    line = line.replace("',", '",')
    line = line.replace(", '", ', "')
    line = line.replace("'}", '"}')
    line = line.replace("\\", "")
    
    try:
        list_of_questions_dicts.append(json.loads(line))
    except Exception as e:
        continue
        #print(e)
        #print line

classifier = GaussianNB()

list_of_questions = []
list_of_labels = []
for i, quest in enumerate(list_of_questions_dicts):
    new_list = []
    for j, word in enumerate(quest['question'].split()):
        new_list.append(word)
    list_of_questions.append(new_list)
    print quest['questionType']
    if quest['questionType'] == "yes/no": list_of_labels.append(1)
    else: list_of_labels.append(0)
        
dictionary = corpora.Dictionary(list_of_questions)
print dictionary.token2id
```

```python
list_of_questions_numerical = []

for i in list_of_questions:
    new_list = []
    for word in i:
        new_list.append(dictionary.token2id[word])

    list_of_questions_numerical.append(new_list)
    
length = len(sorted(list_of_questions_numerical, key=len, reverse=True)[0])
y=np.array([xi+[0]*(length-len(xi)) for xi in list_of_questions_numerical])

print y
    
list_of_labels = np.array(list_of_labels)
print list_of_labels

classifier.fit(y, list_of_labels)
```

```python
import networkx as nx

personal = ['i', 'we']
related_to_this = ['this phone', 'this', 'it', 'the product', 'the phone', 'phone']

wordnet_lemmatizer = WordNetLemmatizer()

cur_dict = {}
rel_list = []
for i, page in enumerate(os.listdir('./pages')):
    if i==1: cur_dict, rel_list = retrieve_info_from_page(page)
        
G = nx.Graph()

origin_node = cur_dict['asin'][:-5]

G.add_node(origin_node)
for i in cur_dict:
    G.add_edge(origin_node, i)
    G.add_edge(i, cur_dict[i])
    
for rel_dict in rel_list:
    for i in rel_dict: 
        #print rel_dict['SBJ']
        if 'SBJ' in rel_dict[i] and 'OBJ' in rel_dict[i] and 'VP' in rel_dict[i]:
            
            sbj = rel_dict[i]['SBJ'].string.lower()
            vp = wordnet_lemmatizer.lemmatize(rel_dict[i]['VP'].string.lower())
            obj = rel_dict[i]['OBJ'].string.lower()
            
            if sbj in personal: continue
            
            if (sbj in related_to_this):
                G.add_edge(origin_node, vp)
                G.add_edge(vp, obj)
            #else:
                #G.add_edge(sbj, vp)
                #G.add_edge(vp, obj)
 
list_of_verbs = set()
for rel_dict in rel_list:
    for i in rel_dict:
        vp = wordnet_lemmatizer.lemmatize(rel_dict[i]['VP'].string.lower())
        list_of_verbs.add(vp)
        #if vp in G.nodes:
        #    if len(G[vp])<=2: G.remove_node(vp)
```
